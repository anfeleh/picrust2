```python
# remove env
!conda env remove -n picrust2_env -y || true

# channels (order matters)
!conda config --remove-key channels || true
!conda config --set channel_priority strict
!conda config --add channels conda-forge
!conda config --add channels bioconda

# create env with python 3.8 (required for picrust2 2.5.2)
!conda create -y -n picrust2_env python=3.8

# install mamba in base (faster solver)
!conda install -y -n base -c conda-forge mamba

# install picrust2 using mamba (solves biom-format better)
!mamba install -y -n picrust2_env -c conda-forge -c bioconda picrust2=2.5.2

```
